```python
# Importeren van de benodigde packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Lees de gegevens in vanuit een CSV-bestand
data = pd.read_csv('perfect_max_spanning.CSV', delimiter=';')

# Filter van 1 piek door tijds-as te knippen
data = data.loc[(data['tijd'] >= -0.00004) & (data['tijd'] <= 0.00001)]

# Plaats de tijd en de spanning in aparte variabelen
tijd = data['tijd']
spanning = data['Spanning']
    
# Plot de gegevens
plt.plot(tijd, spanning)
plt.xlabel('$Tijd$ [s]')
plt.ylabel('$Spanning$ [V]')
plt.title('Spanning over de weerstand')
plt.show()
```

```python
# Bepaal de maximale spanning
max_spanning = np.max(spanning)

# Bepaal de spanning bij 63,2% en 36,8% van de maximale spanning
v1 = 0.632 * max_spanning
v2 = 0.368 * max_spanning

# Zoek de indices van de spanningen het dichtst bij max_spanning en v2
i1 = data.loc[data[data['Spanning'] == max_spanning].index[0], 'tijd']
i2 = data.loc[(data['Spanning']-v2).abs().idxmin(), 'tijd']

# Bepaal de tijd die nodig is om de spanning te veranderen tussen max_spanning en v2
tijdconstante = i2-i1

# Print de maximalen spanning en de tijdsconstante
print("De maximalen spanningen is:", max_spanning, "V")
print("De tijdsconstante is:", round(tijdconstante,9), "s")
```
