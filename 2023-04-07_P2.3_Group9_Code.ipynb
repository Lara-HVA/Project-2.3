```python
#Importeren van de packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importeren van de data
data = np.loadtxt('F0001CH1.CSV', delimiter=';', dtype=np.float32)

#Het schijden van de data
tijd = data[:,0]
spanning = data[:,1]
```

```python
#Het plotten van de data
plt.plot(tijd, spanning)
```

```python
#Functie die pieken telt
def count_peaks(data, threshold):
    peaks = 0
    for i in range(1, len(data)-1):
        if data[i] > data[i-1] and data[i] > data[i+1] and (data[i] - max(data[i-1], data[i+1])) > threshold:
            peaks += 1
    return peaks
```

```python
#Het berekenen en printen van de pieken 
pieken = count_peaks(spanning, 0.02)
print(pieken)
```
